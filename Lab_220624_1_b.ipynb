```python
# importing the required libraries

import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

from getpass import getpass 
```

```python
# crating a connection with the SQL server (sakila database)

password = getpass()
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'  
engine = create_engine(connection_string)
```

```python
# Calculating nr of different actors(last names)?


# importing the data  from the table "Actor"

data = pd.read_sql_query('SELECT * FROM actor', engine) 
data.head()
```

```python


query = '''SELECT DISTINCT last_name
FROM sakila.actor'''
data = pd.read_sql_query(query, engine)
len(data)


# there are 121 actors with different last names
```

```python
# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.


# We just need to make the proper filter while importing the data from the SQL table "rental"

query = '''SELECT *, 
CASE 
WHEN weekday(rental_date) between 0 and 5 then 'Workday'
ELSE 'Weekend'
END AS day_of_week
from sakila.rental;'''

data = pd.read_sql_query(query, engine)
data.head()
```

```python
# Get all films with ARMAGEDDON in the title.

# importing the data from the table "film". The filter is done using SQL

query = """SELECT *
FROM sakila.film;"""
data = pd.read_sql_query(query, engine)
data

# filtering the dataset in order to only get the rows with Armageddon in the column "Title"

data[data['title'].apply(lambda row: 'ARMAGEDDON' in row)]
```

```python
# Get 10 the longest films.


# importing the data by descender order, and taking 10 as limit

query = '''SELECT *
FROM sakila.film
ORDER BY length DESC      
LIMIT 10;'''
data = pd.read_sql_query(query, engine)

data
```

```python
# How many films include Behind the Scenes content?

#if the film has "Behind the Scenes content", this string is mentioned in the column "special_features" (table "film")
```

```python
# importing the table "film" from sakila SQL database

query = '''SELECT * 
FROM sakila.film;'''
data = pd.read_sql_query(query, engine)

# filtering the imported dataset using the lambda function onver the column "special_features"

data_bs= data[data['special_features'].apply(lambda row: 'Behind the Scenes' in row)]

data_bs

```

```python
data_bs.shape    # there are 538 films with "Behind the Scenes content"
```

```python
# Which kind of movies (rating) have a mean duration of more than two hours?
```

```python

# using the SQL command "Having" while making the import in order to filter the table (with "Group by" each rating )


query = '''SELECT rating, round(AVG(length), 2) AS mean_length
FROM sakila.film
GROUP BY rating
HAVING mean_length > 120;'''
data = pd.read_sql_query(query, engine)
data
```
